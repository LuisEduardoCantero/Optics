# Codigo para calcular α y β del vidrio

```python
# Aqui cargamos las liberias a usar
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.integrate import quad
from scipy import optimize
```

### Declaramos los parametros para calcular α y β

```python
w = np.longdouble(np.arange(10**14,10**15,1000000000000))
mu = 1
sigma = np.longdouble(10**-4)                     ## S/m     conductividad electrica
epsilon_vacio = np.longdouble(8.8541878176*10**(-12))  ## F*m-1   permitividad electrica en el vacio
epsilon_rel = 7                                   ##         permitividad electrica relativa
epsilon = epsilon_vacio * epsilon_rel
```

### Calculamos α y β

```python
alfa = np.longdouble(w * np.sqrt(mu*epsilon/2*(np.sqrt(1+sigma**2/(np.power(w,2)*epsilon**2))-1)))
beta = np.longdouble(w * np.sqrt(mu*epsilon/2*(np.sqrt(1+sigma**2/(np.power(w,2)*epsilon**2))+1)))
```

### Graficamos α y β

```python
plt.plot(w, alfa, "-", color = 'red', lw = 4, label="α vidrio")
plt.xlabel("Frequency, w (Hz)")
plt.ylabel("α")
plt.rcParams ["figure.figsize"] = (10,6)
plt.rcParams.update({'font.size':15})
plt.legend()
```

```python
plt.plot(w, beta, "-", color = 'blue', lw = 4, label="β vidrio")
plt.xlabel("Frequency, w (Hz)")
plt.ylabel("β")
plt.rcParams ["figure.figsize"] = (10,6)
plt.rcParams.update({'font.size':15})
plt.legend()
```
