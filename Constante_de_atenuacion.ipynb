# Codigo para graficar una onda E(+z,t) con una cte de atenuacion.

```python
# Aqui cargamos las liberias a usar
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from numpy.fft import fft, fftfreq, ifft
import math
from ipykernel import kernelapp as app
```

### Declaramos variables para una primer onda del campo E atenuandose

```python
z = np.arange(0.00,0.5,0.001)
t = .1
delta_t = t + 0.1

E0 = 10 
alfa = 10 
w= 10 
beta = 100 

campo_e = E0 * np.exp(-alfa * z) * np.cos(w * t - beta * z)
campo_e2 = E0 * np.exp(-alfa * z) * np.cos(w * delta_t - beta * z)
```

### Graficamos las ondas en la misma figura para ver que tienen una diferencia en su frecuencia unicamente

```python
plt.plot(z, campo_e, "-", color = 'red', lw = 4, label="Campo E(+z, t)")
plt.plot(z, campo_e2, "-", color = 'blue', lw = 4, label="Campo E(+z, $\Delta$t)")
plt.xlabel("z")
plt.ylabel("E(z, t)")
plt.ylim(-10, 10)
plt.rcParams ["figure.figsize"] = (10,8)
plt.rcParams.update({'font.size':15})
plt.legend()
```
