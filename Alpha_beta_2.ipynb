# Codigo para calcular α y β de conductor, dielectrico con y sin perdidas.

```python
# Aqui cargamos las liberias a usar
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.integrate import quad
from scipy import optimize
```

### Calculamos y graficamos α y β para conductores (agua desionizada)

##### Nota:  α = β en conductores

```python
w = np.longdouble(np.arange(10**14,10**15,1000000000000))
mu = 1
sigma = np.longdouble(5.5*10**-6)                   ## S/m     conductividad electrica
epsilon_vacio = np.longdouble(8.8541878176*10**(-12))## F*m-1   permitividad electrica en el vacio
epsilon_rel = 80                                   ##         permitividad electrica relativa
epsilon = epsilon_vacio * epsilon_rel
beta = np.longdouble(np.sqrt(w * mu * sigma / 2))

plt.plot(w, beta, "-", color = 'blue', lw = 4, label="α, β agua desionizada")
plt.xlabel("Frequency, w (Hz)")
plt.ylabel("α, β")
plt.rcParams ["figure.figsize"] = (10,4)
plt.rcParams.update({'font.size':15})
plt.legend()
```

### Calculamos β para espacio libre

```python
w = np.longdouble(np.arange(10**14,10**15,1000000000000))
mu = 1
sigma = np.longdouble(0.0)                        ## S/m     conductividad electrica
epsilon_vacio = np.longdouble(8.8541878176*10**(-12))## F*m-1   permitividad electrica en el vacio
epsilon_rel = 1                                   ##         permitividad electrica relativa
epsilon = epsilon_vacio * epsilon_rel
beta = np.longdouble(w * np.sqrt(mu * epsilon))

plt.plot(w, beta, "-", color = 'blue', lw = 4, label="β Madera")
plt.xlabel("Frequency, w (Hz)")
plt.ylabel("β")
plt.rcParams ["figure.figsize"] = (10,4)
plt.rcParams.update({'font.size':15})
plt.legend()
```

### Calculamos β para dielectrico sin perdidas (madera)

```python
w = np.longdouble(np.arange(10**14,10**15,1000000000000))
mu = 1
sigma = np.longdouble(0.0)                        ## S/m     conductividad electrica
epsilon_vacio = np.longdouble(8.8541878176*10**(-12))## F*m-1   permitividad electrica en el vacio
epsilon_rel = 5                                   ##         permitividad electrica relativa
epsilon = epsilon_vacio * epsilon_rel
beta = np.longdouble(w * np.sqrt(mu * epsilon))

plt.plot(w, beta, "-", color = 'green', lw = 4, label="β Madera")
plt.xlabel("Frequency, w (Hz)")
plt.ylabel("β")
plt.rcParams ["figure.figsize"] = (10,4)
plt.rcParams.update({'font.size':15})
plt.legend()
```
