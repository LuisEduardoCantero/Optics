# Codigo para graficar la ecuacion de Cauchy y Sellmeier

```python
# Aqui cargamos las liberias a usar
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.integrate import quad
from scipy import optimize
```

### Declaramos los parametros para el BK7 y graficamos Cauchy y Sellmeier

```python
l_onda =  np.longdouble(np.arange(0.2,1.6,0.01))

A = 1.5046
B = 0.00420
n = A + B /(l_onda**2) ## Cauchy

B1 = 1.03961212 
B2 = 0.231792344
B3 = 1.01046945
C1 = 6.00069867e-3
C2 = 2.00179144e-2 
C3 = 1.03560653e2

n2 =  np.longdouble(np.sqrt(1 + (B1*l_onda**2)/(l_onda**2-C1) + (B2*l_onda**2)/(l_onda**2-C2) + (B3*l_onda**2)/(l_onda**2-C3)))

plt.plot(l_onda, n, "-", color = 'blue', lw = 4, label="BK7 Cauchy")
plt.plot(l_onda, n2, "--", color = 'green', lw = 4, label="BK7 Sellmeier")
plt.xlabel("Longitud de onda ($\mu$m)")
plt.ylabel("n(Î»)")
plt.rcParams ["figure.figsize"] = (10,6)
plt.rcParams.update({'font.size':15})
plt.legend()
```
