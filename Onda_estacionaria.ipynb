## Codigo para una onda que no viaja

```python
# Aqui cargamos las liberias a usar
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.integrate import quad
from scipy import optimize
```

### Declaramos algunas variables para $E_1$ y $H_1$

```python
Eio = 1
beta1 = 5
z =   np.arange(-10,0.0,0.01)
z2 =   np.zeros(len(z))
z3 =   np.arange(0,7,0.01)
w = 1.0
t1 = 0
t2 = 2 * 3.1415926 / w  / 8
t3 = 2 * 3.1415926 / w  / 4
t4 = 2 * 3.1415926 / w * 3 / 8
t5 = 2 * 3.1415926 / w * 1 / 2

E1 = 2 * Eio * np.sin(beta1 * z) * np.sin(w * t1)
E2 = 2 * Eio * np.sin(beta1 * z) * np.sin(w * t2)
E3 = 2 * Eio * np.sin(beta1 * z) * np.sin(w * t3)
E4 = 2 * Eio * np.sin(beta1 * z) * np.sin(w * t4)
E5 = 2 * Eio * np.sin(beta1 * z) * np.sin(w * t5)
Eb = 2 * Eio * np.sin(beta1 * z) * np.sin(w * t3)
Eb2 = 10 * np.sin(1/z3 * z3)
```

```python
plt.plot(z, E1, "-", color = 'blue', lw = 2, label="$E_1$")
plt.plot(z, E2, "-", color = 'gray', lw = 2, label="$E_2$")
plt.plot(z, E3, "-", color = 'red', lw = 2, label="$E_3$")
plt.plot(z, E4, "-", color = 'green', lw = 2, label="$E_4$")
plt.plot(z, E5, "-", color = 'purple', lw = 2, label="$E_5$")
plt.plot(z2, Eb, "-", color = 'black', lw = 4)
plt.plot(z3, Eb2, "-", color = 'black', lw = 4)


plt.xlabel("z")
plt.ylabel("x")
plt.ylim(-2.2,2.2)
plt.rcParams ["figure.figsize"] = (10,6)
plt.rcParams.update({'font.size':15})
plt.legend()
```

```python
Eio = 0.1
beta1 = 5
z =   np.arange(-10,0.0,0.01)
z2 =   np.zeros(len(z))
z3 =   np.arange(0,7,0.01)
n = 0.1
w = 1.0
t1 = 0
t2 = 2 * 3.1415926 / w  / 8
t3 = 2 * 3.1415926 / w  / 4
t4 = 2 * 3.1415926 / w * 3 / 8
t5 = 2 * 3.1415926 / w * 1 / 2

H1 = 2 * Eio / n * np.cos(beta1 * z) * np.cos(w * t1)
H2 = 2 * Eio / n * np.cos(beta1 * z) * np.cos(w * t2)
H3 = 2 * Eio / n * np.cos(beta1 * z) * np.cos(w * t3)
H4 = 2 * Eio / n * np.cos(beta1 * z) * np.cos(w * t4)
H5 = 2 * Eio / n * np.cos(beta1 * z) * np.cos(w * t5)

Eb = 20 * Eio * np.sin(beta1 * z) * np.sin(w * t3)
Eb2 = 10 * np.sin(1/z3 * z3)
```
