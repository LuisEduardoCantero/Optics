# Codigo para generar pulsaciones

```python
# Aqui cargamos las liberias a usar
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from numpy.fft import fft, fftfreq, ifft
import math
from ipykernel import kernelapp as app
```

### Declaramos variables para una primer onda

$y_1(x,t) = A * cos(w*t-k*x)$

```python
x = np.arange(0.01,100,0.01) # declaramos un arreglo de numeros para la posicion x de la onda
amp = 5 # Valor de la amplitud
w = 2 # valor de la velocidad angular
k= 2 # Valor del numero de onda
t=1 # Tiempo
y1 = amp*np.cos(w*t - k*x) # calculamos la funcion de onda en el intervalo xe
```

### Declaramos variables para una segunda onda

$y_2(x,t) = A * cos((w + \Delta w)*t-(k+ \Delta k)*x)$

```python
dw = 2 # Valor del desfase de la velocidad angular
dk = 0.1 # Valor del desfase del numero de onda
y2 = amp*np.cos((w + dw)*t + (k + dk)*x) # calculamos la funcion de onda en elintervalo x establecido
```

### Graficamos las ondas en la misma figura para ver que tienen una diferencia en su frecuencia unicamente

```python
plt.plot(x,y1, "-", color = 'red', lw = 4, label="Onda 1")
plt.plot(x,y2, "-", color = 'blue', lw = 4, label="Onda 2")
plt.xlabel("x")
plt.ylabel("y(x, t)")
plt.ylim(-10, 10)
plt.rcParams ["figure.figsize"] = (30,10)
plt.rcParams.update({'font.size':25})
plt.legend()
```

### Obtenemos el beat (pulso) producido por la superposicion de ambas ondas 

$y_3(x,t) = 2 * A * cos(w*t - k*x)*cos(\frac{\Delta w}{2}*t - \frac{\Delta k}{2}*x)$

```python
y3 = 2 * amp * np.cos(w*t-k*x)*np.cos(dw/2*t - dk/2*x)
```

### Graficamos la superposicion de las ondas

```python
plt.plot(x,y3, "-", color = 'purple', lw = 5, label="Onda 1 + Onda 2")
plt.xlabel("x")
plt.ylabel("y(x, t)")
plt.ylim(-15, 15)
plt.rcParams ["figure.figsize"] = (30,10)
plt.rcParams.update({'font.size':30})
plt.legend()
```
